```python
# Import Libraries
import string, math
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline


# Function to get and divide dataset into y(Blood pressure) and X (Age)
def preprocess():
    f = open("grades_data.txt", "r")
    if f.mode == 'r':
        # Read dataset
        contents = f.read()
        dataset = np.array([[float(j) for j in i.split(',')] for i in contents.splitlines()])

        # Separate features from results 
        X, y = dataset[:, :-1], dataset[:, 2:]
        y = y.astype(int)
        print("Logistic Regression Algorithm")
        print("Red 'x' means student failed, and green 'o' means student was approved: ")
        
        # Plotting the 
        plt.title("Full dataset student results")
        plt.xlabel('Exam 1 score')
        plt.ylabel('Exam 2 score')
        for i in range(len(X)):
            if y[i] == 1:
                plt.plot(X[i,0],X[i,1], 'go')
            else:
                plt.plot(X[i,0],X[i,1], 'rx')
        plt.axis([min(X[:,0]), 100, min(X[:,1]), 100])
        plt.show()
        
        # Divide dataset into test set (70%) and train set (30%)
        training_set, testing_set = np.split(X,[int(0.7 * len(X))])
        print(len(X))
        print("Training set")
        print(training_set)
        print("Testing set")
        print(testing_set)
        return X, y

# Function to make the Gradient Descent operation
def gradientDescent(y, X, y_data):
    theta1 = 0
    theta2 = 0
    for i in range(1, len(X)):
        theta1 += y(X[i]) - y_data[i]
        theta2 += (y(X[i]) - y_data[i]) * X[i]
    
    return theta1 / len(X), theta2 / len(X)
    
# Function to plot the line between the data
def plot_line(y, data_points):
    # Take min point and max point for the X axis in the dataset
    x_values = [i for i in range(int(np.amin(data_points))-1, int(np.amax(data_points))+2)]
    # For y values use the linnear equation
    y_values = [y(x) for x in x_values]
    plt.plot(x_values, y_values, 'r')
    
# Start App
if __name__ == '__main__':
    # Get data
    X, y_data = preprocess()

    # Start slope in 0
    m = 0

    # Start Y interception in 0
    b = min(y_data)
    # Define linnear equation 
    y = lambda x : 1/(1+np.exp(-x))
    # Set the learning rate
    learning_rate = 0.0000005

    # Gradient Descent logic
    for i in range(50):
        s1, s2 = gradientDescent(y, X, y_data)
        m = m - learning_rate * s2
        b = b - learning_rate * s1
        # Print line between the data on each iteration (to show learnig rate)
#         plot_line(y, X)
#         plt.plot(X, y_data, 'bo')
    
    # Slope
    print("Slope is:", m[0])
    # Y interception in 0
    print("Y intersection is: ", b[0])
    lol = y(54.0246928153624)
    print("prediction is: ", lol)
```
